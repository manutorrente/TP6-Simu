```python
b = 15
m = 0.1
data = []

for i in range(100):
    model = modelo.Modelo(b, m+0.01*i, 1000)
    counter = 0
    for j in range(10):
        model.reset()
        model.init_random()
        counter += model.simular()[0]
    data.append((i, counter/10))
    
```

```python
import matplotlib.pyplot as plt

# Example list of tuples

# Extract x and y values from the list of tuples
x_values = [item[0] for item in data]
y_values = [item[1] for item in data]

# Plot the data
plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.xlabel('Number of Loops')
plt.ylabel('Result')
plt.title('Plot of Results vs. Number of Loops')
plt.grid(True)
plt.show()
```
