```python
import matplotlib.pyplot as plt
from scipy.stats import uniform, genpareto, johnsonsb

class DistribucionesEasyFit:
    def generar_proxima_llegada(self):
        return uniform.rvs(loc=1.3675, scale=5.3525)/7 

    def generar_monto_honorarios(self):
        return genpareto.rvs(c=0.03876, scale=6.5469, loc=20.189)
    
    def generar_duracion_consulta(self):
        return johnsonsb.rvs(a=0.35316, b=0.47835, scale=92.976, loc= -0.88488)*4
distribuciones_easy_fit = DistribucionesEasyFit()
```

```python
# Generate samples for each distribution
llegadas = [distribuciones_easy_fit.generar_proxima_llegada() for _ in range(1000)]


# Plot histogram for 'próxima llegada'
# plt.figure(figsize=(8, 6))
plt.hist(llegadas, bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma de próxima llegada')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
montos = [distribuciones_easy_fit.generar_monto_honorarios() for _ in range(1000)]
plt.hist(montos, bins=30, color='lightgreen', edgecolor='black')
plt.title('Histograma de monto de honorarios')
plt.xlabel('Monto')
plt.ylabel('Frecuencia')
plt.show()
```

```python
duracion = [distribuciones_easy_fit.generar_duracion_consulta() for _ in range(1000)]
plt.hist(duracion, bins=30, color='salmon', edgecolor='black')
plt.title('Histograma de duración de consulta')
plt.xlabel('Duración')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
